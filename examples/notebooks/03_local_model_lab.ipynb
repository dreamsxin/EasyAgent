# Local-model lab

Start with the offline mock provider, then switch the same code to Ollama or DeepSeek. The public Agent API does not change.

```python
import os

from agentmold import Agent

# Set EASYAGENT_MODEL=ollama/llama3 when a local Ollama server is ready.
model = os.getenv("EASYAGENT_MODEL", "mock")
llm = {"provider": "mock", "model": "mock", "temperature": 0} if model == "mock" else model
agent = Agent(
    name="Local Lab",
    instructions="Answer briefly and state uncertainty.",
    llm=llm,
)

print(f"Using: {model}")
print(agent("What makes a good reproducible experiment?"))
agent.last_trace.to_jsonl("runs/local-model.jsonl")
```

```python
# Optional hosted configuration. No request is made by this cell.
deepseek_config = {
    "provider": "deepseek",
    "model": "deepseek-v4-flash",
    "api_key": os.getenv("DEEPSEEK_API_KEY"),
    "base_url": "https://api.deepseek.com",
}
if deepseek_config["api_key"]:
    hosted_agent = Agent(llm=deepseek_config)
    print("DeepSeek configuration is ready.")
else:
    print("Set DEEPSEEK_API_KEY to create the hosted agent.")
```

The trace file records the model name, timing, messages, and final answer without putting API keys into the artifact.
