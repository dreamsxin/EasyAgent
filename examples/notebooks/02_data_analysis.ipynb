# Data analysis lab

Keep deterministic data preparation in Python and use the Agent for the natural-language part. This separation makes the result easy to test and reproduce.

```python
from statistics import mean

from agentmold import Agent, EvalCase, evaluate

sales = [
    {"region": "north", "revenue": 120},
    {"region": "north", "revenue": 180},
    {"region": "south", "revenue": 90},
    {"region": "south", "revenue": 150},
]

by_region = {}
for row in sales:
    by_region.setdefault(row["region"], []).append(row["revenue"])
summary = {region: {"mean": mean(values), "n": len(values)} for region, values in by_region.items()}
summary
```

```python
def build_agent() -> Agent:
    return Agent(
        name="Data Analyst",
        instructions="Explain the supplied data without inventing numbers.",
        llm="mock",
    )

prompt = f"Summarize this regional revenue table: {summary}"
report = evaluate(
    build_agent,
    [EvalCase(name="regional-summary", input=prompt, expected=f"[mock-llm] {prompt}")],
)
print(report.mean_score, report.passed, report.failed)
report.to_json("runs/data-analysis-report.json")
```

The same `evaluate` call can use a hosted model, a custom scorer, and multiple cases. Keep the dataset and scorer in version control so a prompt change has a measurable result.
