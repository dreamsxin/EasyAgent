# Literature review lab

Build a small literature assistant from ordinary Python functions. The notebook is offline-first: `mock` makes it runnable in class or CI, while the same agent can later use DeepSeek by changing one argument.

```python
from agentmold import Agent, Memory, tool

PAPERS = [
    {"title": "Retrieval-Augmented Generation for Knowledge-Intensive NLP Tasks", "year": 2020},
    {"title": "Self-Refine: Iterative Refinement with Self-Feedback", "year": 2023},
    {"title": "Toolformer: Language Models Can Teach Themselves to Use Tools", "year": 2023},
]

@tool
def search_papers(query: str) -> str:
    """Find papers whose title overlaps the query.

    Args:
        query: Research terms to search for.
    """
    terms = set(query.lower().split())
    matches = [
        paper for paper in PAPERS
        if terms & set(paper["title"].lower().split())
    ] or PAPERS[:2]
    return "\n".join(f"{p['year']}: {p['title']}" for p in matches)
```

```python
agent = Agent(
    name="Literature Reviewer",
    instructions="Find relevant papers, then explain the evidence clearly.",
    tools=[search_papers],
    memory=Memory(max_messages=20),
    llm="mock",
)

# The mock model recognizes `tool:` and exercises the real tool loop offline.
answer = agent("tool: find papers about retrieval and tool use")
print(answer)

# Every run can be saved for later comparison.
agent.last_trace.to_jsonl("runs/literature-review.jsonl")
```

## Use a hosted model

Replace `llm="mock"` with `llm="deepseek/deepseek-v4-flash"` after setting `DEEPSEEK_API_KEY`. Keep the tool and memory code unchanged.
